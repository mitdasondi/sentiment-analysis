```
from textblob import TextBlob
def find_sentiment(text):
    if TextBlob(text).sentiment.polarity < 0.5:
        return "sentiments are very negative"
    elif TextBlob(text).sentiment.polarity > 0 :
        return "sentiments are positive"
    elif TextBlob(text).sentiment.polarity > 0 and TextBlob(text).sentiment.polarity < 0.5:
        return "sentiments are positive"

    elif TextBlob(text).sentiment.polarity > 0.5:
        return "very negative"
    else: return "natural"

x = ""

print (find_sentiment(x))
```

```
from textblob import TextBlob

def find_sentiment(text):
    if TextBlob(text).sentiment.polarity < 0.5:
        return "sentiments are very negative"
    elif TextBlob(text).sentiment.polarity > 0 :
        return "sentiments are positive"
    elif TextBlob(text).sentiment.polarity > 0 and TextBlob(text).sentiment.polarity < 0.5:
        return "sentiments are positive"

    elif TextBlob(text).sentiment.polarity > 0.5:
        return "very negative"
    else: return "neutral"

import pandas as pd

df = pd.read_csv('test_oJQbWVk.csv')
df['sentiment'] = df['tweet'].apply(find_sentiment)


print(df[['tweet','sentiment']])
df.to_csv()
```
